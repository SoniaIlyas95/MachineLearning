# Import libaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import online data

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

# Input variables vs features

```python
digits.data.shape
X = digits.data
```

# Outlables

```python
digits.target.shape
y = digits.target
```

# Plotting

```python
plt.figure(figsize=(20, 4))
for index, (Image, label) in enumerate(zip(digits.data[0: 10], digits.target[0: 10])):
    plt.subplot(1,10, index + 1)
    plt.imshow(np.reshape(Image, (8,8)), cmap = plt.cm.gray)
    plt.title('Training: %i/n' % label, fontsize= 20)
```

```python
plt.figure(figsize=(20, 4))
for index, (Image, label) in enumerate(zip(digits.data[0: 5], digits.target[0: 5])):
    plt.subplot(1,5, index + 1)
    plt.imshow(np.reshape(Image, (8,8)), cmap = plt.cm.gray)
    plt.title('Training: %i/n' % label, fontsize= 20)
```

# Split the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)
```

# Shape of dataset

```python
print('Train input data:', X_train.shape)
print('Test input data:', X_test.shape)
print('Train output data:', y_train.shape)
print('Test output data:', y_test.shape)
```

# Model train

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X_train, y_train)
model
```

# Prediction

```python
model.predict(X_test[0:11])
predictions= model.predict(X_test)
predictions
```

# Accuracy prediction

```python
score = model.score(X_test, y_test)
print("The accuracy score is:", score)
```

# Confusion metrics

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predictions)
cm
```

```python
import seaborn as sns
plt. figure(figsize=(12, 12))
sns.heatmap(cm, annot= True, fmt=".3f", linewidths = .5, square = True, cmap = 'Spectral')
plt.ylabel('Actual lable')
plt.xlabel('Predicted lable')
all_sample_title ='Accuracy Score: {0}', format(score*100, '%')
plt.title(all_sample_title, size = 15)
```

# Getting misclassifier labels

```python
import numpy as np
import matplotlib.pyplot as plt
index = 0
misclassifiedindexes = []
for label, predict in zip(y_test, predictions):
    if label != predict:
        misclassifiedindexes.append(index)
        index +=1
```

# Plotting misclassified labels with known lables

```python
plt.figure(figsize= (20, 4))
for plotindex, badindex in enumerate(misclassifiedindexes[0:5]):
    plt.subplot(1, 5, plotindex + 1)
    plt.imshow(np.reshape(X_test[badindex], (8,8)), cmap=plt.cm.gray)
    plt.title("Predicted: {}, Actual: {}".format(predictions[badindex], y_test[badindex]), fontsize=15)
```

```python
# plotting misclassified labels with known lables
plt.figure(figsize= (20, 4))
for plotindex, badindex in enumerate(misclassifiedindexes[11:21]):
    plt.subplot(3, 13, plotindex + 1)
    plt.imshow(np.reshape(X_test[badindex], (8,8)), cmap=plt.cm.gray)
    plt.title("Predicted: {}, Actual: {}".format(predictions[badindex], y_test[badindex]), fontsize=15)
```
